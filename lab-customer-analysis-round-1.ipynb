## 7. Data Cleaning 

```python
# drop education and number of complaints columns
file = file.drop(columns=['education', 'numb_open_complaints'])
```

```python

duplicates_number = file[['id', 'state', 'gender', 'customer_ltv', 'income','monthly_premium_auto', 'total_claim_amount','policy_type', 'vehicle_class']].duplicated()
print("The duplicate rows :", duplicates_number.sum())
```

```python
file.info()
```

```python
file.dropna(subset=['id'], inplace=True)
```

```python
file.isna().sum()
```

```python
file['gender'].value_counts()
```

```python
# Define a dictionary to map the inconsistent values to the correct ones
replacement_dict = {'F': 'Female', 'M': 'Male', 'Femal': 'Female', 'female' : 'Female'}

# Replace the inconsistent values in the 'gender' column using the dictionary
file['gender'] = file['gender'].replace(replacement_dict)
```

```python
# Replacing null values for categorical variables
file['gender'] = file['gender'].fillna('Other')
```

```python
file['gender'].value_counts()
```

```python
file['customer_ltv'].isna().sum()
```

```python
#Correct the values in the column customer lifetime value. They are given as a percent, so multiply them by 100 and change dtype to numerical type.
file["customer_ltv"]=file["customer_ltv"].astype(str).str.replace("%", "").astype(float)
```

```python
file['customer_ltv'] = file['customer_ltv'] * 100
```

```python
file.head()
```

```python
col_median = file["customer_ltv"].median()
file["customer_ltv"] = file["customer_ltv"].fillna(col_median)
```
